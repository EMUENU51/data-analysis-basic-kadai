```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, tight_layout=True)

x = [0, 1, 2, 3]
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)

x = np.linspace(0, 5, 50)
y = 2 + 2 * np.sin(2 * x)
axes[0, 1].plot(x, y)

x = np.random .normal(10, 5, 1000)
axes[0, 2].hist(x)

np.random.seed(3)
x = 4 + np.random.normal(0, 2, 60)
y = 4 + np.random.normal(0, 2, len(x))
axes[1, 0].scatter(x, y)

plt.show()
```
